# **1. Classification**

```python
clas = tree.DecisionTreeClassifier(max_depth=2)
iris = load_iris()

X_train = iris.data
y_train = iris.target
clas.fit(X_train, y_train)

# from sklearn.tree import plot_tree
# plot_tree(clas)


viz = dtreeviz.model(clas,
               X_train,
               y_train,
               feature_names=iris.feature_names,
               class_names=["setosa", "versicolor", "virginica"])
viz.view()
```

# **2. Regression**

```python
regr = tree.DecisionTreeRegressor(max_depth=2)
diabities = load_diabetes()

X_train = diabities.data
y_train = diabities.target
regr.fit(X_train,y_train)

viz = dtreeviz.model(regr,
                     X_train,
                     y_train,
                     feature_names=diabities.feature_names,
                     target_name="disease"
                     )
viz.view()
```

# **3. Show Node Labels**

```python
clas = tree.DecisionTreeClassifier(max_depth=2)
iris = load_iris()

X_train = iris.data
y_train = iris.target
clas.fit(X_train, y_train)

X = iris.data[np.random.randint(0,len(iris.data)),:]
# from sklearn.tree import plot_tree
# plot_tree(clas)


viz = dtreeviz.model(clas,
               X_train,
               y_train,
               feature_names=iris.feature_names,
               class_names=["setosa", "versicolor", "virginica"],
                     tree_index=X)
viz.view(show_node_labels=True)
```
