```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
X , y = make_regression(n_samples=4 , n_features = 1 , n_informative=1 , n_targets=1,noise = 80 , random_state = 13)
```

```python
X
```

```python
y
```

```python
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color = "red")
```

# **With Gradient Descent**

```python
# let Slope m = 78.35
# let starting point of b = 0
y_pred = ((78.35 * X) + 0).reshape(4)
y_pred
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color = "red", label = "OLS")
plt.plot(X,y_pred,color = "blue", label = "b = 0")
plt.legend()
plt.show()
```

```python
m = 78.35
b = 0

loss_slope = -2 * np.sum(y - m * X.ravel() - b)
loss_slope
```

```python
# let learning rate  = 0.1
lr = 0.1

step_size = loss_slope * lr
step_size
```

```python
## Value of New B is

b = b - step_size
b
```

```python
y_pred1 = ((78.35 * X) + b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color = "red", label = "OLS")
plt.plot(X,y_pred,color = "blue", label = "b = 0")
plt.plot(X,y_pred1 , color = "green",label = f"b = {round(b,3)}")
plt.legend()
plt.show()
```

```python
## Iteration 2
loss_slope = -2 * np.sum(y - m * X.ravel() -b )
loss_slope
```

```python
step_size = loss_slope * lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred2 = ((78.35 * X) + b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color = "red", label = "OLS")
plt.plot(X,y_pred,color = "blue", label = "b = 0")
plt.plot(X,y_pred1 , color = "green",label = f"b = Iteration 1")
plt.plot(X,y_pred2 , color = "orange",label = f"b = Iteration 2")
plt.legend()
plt.show()
```

```python
# Iteration 3

loss_slope = -2 * np.sum(y - m * X.ravel() -b)
step_size = loss_slope * lr
b = b - step_size
y_pred3 = ((78.35 * X) + b).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color = "red", label = "OLS")
plt.plot(X,y_pred,color = "blue", label = "b = 0")
plt.plot(X,y_pred1 , color = "green",label = f"b = Iteration 1")
plt.plot(X,y_pred2 , color = "orange",label = f"b = Iteration 2")
plt.plot(X,y_pred3 , color = "pink",label = "b = Iteration 3")
plt.legend()
plt.show()
```

```python
## With For Loop

b = -100
m = 78.35
lr = 0.01

epochs = 100

for i in range(epochs):
  loss_slope = -2 * np.sum(y - m * X.ravel() -b)
  step_size = loss_slope * lr
  b = b - step_size

  y_pred = m * X + b
  plt.plot(X,y_pred)
plt.scatter(X,y,color = "red",label= "OLS")
plt.legend()
```
