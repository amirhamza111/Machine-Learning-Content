```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("titanic (1).csv")
df.head()
```

```python
df["number"].unique()
```

```python
fig = df["number"].value_counts().plot(kind = "bar")
fig.set_title("Passengers Travelling Alone")
```

```python
## Extract Numerical Part
df["number_numerical"] = pd.to_numeric(df["number"] , errors = "coerce" , downcast = "integer")
df["number_categorical"] = np.where(df["number_numerical"].isnull() , df["number"] , np.nan)
df.head()
```

```python
df["Cabin"].unique()
```

```python
df["Ticket"].unique()
```

```python
df["cabin_num"] = df["Cabin"].str.extract("(\d+)")
df["cabin_cat"] = df["Cabin"].str.extract("([A-Za-z])")
df["number_numerical"] = df["number_numerical"].fillna(0)
df["number_categorical"] = df["number_categorical"].fillna("Missing")
df.sample(10)
```

```python
fig = df["cabin_cat"].value_counts().plot(kind = "bar")
fig.set_title("Cabin Category")
```

```python
df["Ticket"].head()
```

```python
## Extract last part of ticket for number
df["ticket_num"] = df["Ticket"].apply(lambda s : s.split()[-1])
df["ticket_num"] = pd.to_numeric(df["ticket_num"] , errors = "coerce", downcast = "integer")

## Extract first part of ticket for cat
df["ticket_cat"] = df["Ticket"].apply(lambda s : s.split()[0])
df["ticket_cat"] = np.where(df["ticket_cat"].str.isdigit() , np.nan , df["ticket_cat"])
df.head()
```

```python
df["ticket_cat"].value_counts().plot(kind = "bar")
```
