```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("train.csv" , usecols = ["Survived","Age","SibSp","Parch","Pclass"])
df.head()
```

```python
df["Family"] = df["SibSp"] + df["Parch"] + 1
df.sample(10)
```

```python
def family(n):
  if n == 1:
    return 0
  elif n > 1 and n <= 4:
    return 1
  else:
    return 2
```

```python
df["Family_type"] = df["Family"].apply(family)
df.head()
```

```python
df.sample(10)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.impute import SimpleImputer
```

```python
si = SimpleImputer()
X = df.drop(columns = ["Survived","SibSp","Parch","Family"])
y = df["Survived"]
```

```python
X = pd.DataFrame(si.fit_transform(X),columns = X.columns)
```

```python
X.head()
```

```python
X.shape , y.shape
```

```python
from sklearn.model_selection import cross_val_score

lr = LogisticRegression()
cross_val_score(lr, X , y , cv = 10).mean()
```
