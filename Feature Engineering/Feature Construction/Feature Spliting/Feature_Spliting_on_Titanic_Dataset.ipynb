```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("train.csv" , usecols = ["Survived","Age","Name"])
df.head()
```

```python
df["Salutation"] = df["Name"].str.split(",",expand = True)[1].str.split(".",expand = True)[0]
```

```python
df.head()
```

```python
df["Salutation"].value_counts().plot(kind = "bar")
```

```python
df["Salutation"].value_counts().plot(kind = "pie")
```

```python
# prompt: Generate code for normalizing df["Age"]

# Assuming df and necessary libraries are already imported and df["Age"] exists.
# If not, please include the code to load the dataframe and create the "Age" column.

df['Age'] = df['Age'].fillna(df['Age'].mean()) #fill null values with mean
min_age = df["Age"].min()
max_age = df["Age"].max()
df["Age"] = (df["Age"] - min_age)/(max_age - min_age)
```
