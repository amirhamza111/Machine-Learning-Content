```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/wine_data.csv" , header = None , usecols = [0,1,2])
df.columns = ["Class" , "Alcohol" , "Malic Acid"]
df.head()
```

# **Max Absolute Scaling**

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(df.drop("Class" , axis = 1) ,
                                                       df["Class"] ,
                                                       test_size = 0.3 ,
                                                       random_state = 0)
```

```python
from sklearn.preprocessing import MaxAbsScaler
s = MaxAbsScaler()
s.fit(X_train)
X_train_scaled = pd.DataFrame(s.transform(X_train) , columns = X_train.columns)
X_test_scaled = pd.DataFrame(s.transform(X_test) , columns = X_train.columns)
```

```python
fig , (ax1 , ax2) = plt.subplots(ncols = 2 , figsize = (10 , 5))
# Before Scaling
sns.kdeplot(X_train["Alcohol"] , ax = ax1)
# After Scaling
sns.kdeplot(X_train_scaled["Alcohol"] , ax = ax2)
plt.show()
```

```python
fig , (ax1 , ax2) = plt.subplots(ncols = 2 , figsize = (10 , 5))
# Before Scaling
sns.kdeplot(X_train["Malic Acid"] , ax = ax1)
# After Scaling
sns.kdeplot(X_train_scaled["Malic Acid"] , ax = ax2)
plt.show()
```

```python
fig , (ax1 , ax2) = plt.subplots(ncols = 2 , figsize = (10 , 5))
# Before Scaling
ax1.scatter(X_train["Alcohol"] , X_train["Malic Acid"] , c = y_train )
# After Scaling
ax2.scatter(X_train_scaled["Alcohol"] , X_train_scaled["Malic Acid"] , c = y_train )
ax1.legend()
ax2.legend()
plt.show()
```
