```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.array([455,642,433,505,262,641,297,506,387])
y = np.array([45,33,63,65,23,64,78,45,85])
```

```python
### Standardization
def standardization(values):
  mean = np.mean(values)
  std = np.std(values)
  stand = []
  for i in values:
    res = (i-mean)/std
    stand.append(round(res ,4))
  return stand

x_stand = standardization(x)
y_stand = standardization(y)

fig , (ax1 , ax2 ) = plt.subplots(ncols = 2 , figsize = (10,5))
# Before Scaling
ax1.scatter(x , y)
ax1.set_title('Before Scaling')
# After Scaling
ax2.scatter(x_stand , y_stand)
ax2.set_title('After Scaling')
plt.show()
```

```python
## MinMaxScaling
def mms(values):
  min_ = min(values)
  max_ = max(values)
  mms_ = []
  for i in values:
    res = (i-min_)/(max_-min_)
    mms_.append(round(res , 4))
  return mms_

x_mms = mms(x)
y_mms = mms(y)

fig , (ax1 , ax2 ) = plt.subplots(ncols = 2 , figsize = (10,5))
# Before Scaling
ax1.scatter(x , y)
ax1.set_title('Before Scaling')
# After Scaling
ax2.scatter(x_mms , y_mms)
ax2.set_title('After Scaling')
plt.show()
```

```python
### Mean Normalization
def mean_norm(values):
  mean_ = np.mean(values)
  min_ = min(values)
  max_ = max(values)
  mean_norm_ = []
  for i in values:
    res = (i-mean_)/(max_ - min_)
    mean_norm_.append(round(res , 4))
  return mean_norm_

x_mean_norm = mean_norm(x)
y_mean_norm = mean_norm(y)

fig , (ax1 , ax2 ) = plt.subplots(ncols = 2 , figsize = (10,5))
# Before Scaling
ax1.scatter(x , y)
ax1.set_title('Before Scaling')
# After Scaling
ax2.scatter(x_mean_norm , y_mean_norm)
ax2.set_title('After Scaling')
plt.show()
```

```python
### Max Abs Scaling
def max_abs(values):
  abs_max = abs(max(values))
  max_abs = []
  for i in values:
    res = i/abs_max
    max_abs.append(round(res , 4))
  return max_abs

x_max_abs = max_abs(x)
y_max_abs = max_abs(y)

fig , (ax1 , ax2 ) = plt.subplots(ncols = 2 , figsize = (10,5))
# Before Scaling
ax1.scatter(x , y)
ax1.set_title('Before Scaling')
# After Scaling
ax2.scatter(x_max_abs , y_max_abs)
ax2.set_title('After Scaling')
plt.show()
```

```python
## Robust Scaling
def robust(values):
  q1 = np.quantile(values , 0.25)
  q3 = np.quantile(values , 0.75)
  iqr = q3 - q1
  median = np.median(values)
  robust_ = []
  for i in values:
    res = (i-median)/iqr
    robust_.append(round(res ,4))
  return robust_

x_robust = robust(x)
y_robust = robust(y)


fig , (ax1 , ax2 ) = plt.subplots(ncols = 2 , figsize = (10,5))
# Before Scaling
ax1.scatter(x , y)
ax1.set_title('Before Scaling')
# After Scaling
ax2.scatter(x_robust , y_robust)
ax2.set_title('After Scaling')
plt.show()
```
