# **Working with Dates**

```python
# Converting to datetime
date["date"] = pd.to_datetime(date["date"])
date.info()
```

```python
date.drop(columns = ["product_id","city_id"] , axis = 1 , inplace = True)
```

```python
date.drop("orders",axis = 1 , inplace = True)
```

```python
### Check Year
date["date_year"] = date["date"].dt.year

## Check Month
date["date_month"] = date["date"].dt.month

## Check Month Name
date["date_month_name"] = date["date"].dt.month_name()

## Check Day
date["date_day"] = date["date"].dt.day

# Check Day Name
date["date_day_name"] = date["date"].dt.day_name()

## Check Is Last Month
date["date_month_last"] = date["date"].dt.is_month_end

## Check Day of Week
date["date_dow"] = date["date"].dt.dayofweek
date.head()
```

```python
## Is weekend?
date["date_is_weekend"] = np.where(date["date_day_name"].isin(["Sunday","Saturday"]) , 1 , 0)
date.head()
```

```python
## Extract Week of year

date["date_week"] = date["date"].dt.isocalendar().week
date.head()
```

```python
## Extract Quarter
date["Quarter"] = date["date"].dt.quarter
date.head()
```

```python
## Extract Semester

date["Semester"] = np.where(date["Quarter"].isin([1,2]) , 1 , 2)
date.head()
```

```python
##Extract time elapsed between two days

import datetime

today = datetime.datetime.today()
today
```

```python
today-date["date"]
```

```python
(today - date["date"]).dt.days
```
