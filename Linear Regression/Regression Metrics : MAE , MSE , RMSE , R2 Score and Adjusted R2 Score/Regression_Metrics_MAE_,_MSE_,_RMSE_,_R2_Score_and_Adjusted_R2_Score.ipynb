```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv("/content/placement (1).csv")
df.head()
```

```python
plt.scatter(df["cgpa"] , df["package"])
plt.xlabel("cgpa")
plt.ylabel("package")
plt.title("cgpa vs package")
```

```python
from sklearn.model_selection import train_test_split
X_train  , X_test , y_train ,y_test = train_test_split(df[["cgpa"]],df["package"],test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
plt.scatter(df["cgpa"] , df["package"])
plt.plot(X_test,y_pred,color="red")
plt.xlabel("cgpa")
plt.ylabel("package")
plt.title("cgpa vs package")
```

```python
from sklearn.metrics import mean_absolute_error , mean_squared_error , r2_score
```

```python
print("MAE:",round(mean_absolute_error(y_test,y_pred) ,3), "LPA")
```

```python
print("MSE:",round(mean_squared_error(y_test,y_pred) ,3), "LPA**2")
```

```python
print("RMSE:",round(np.sqrt(mean_absolute_error(y_test,y_pred)) ,3), "LPA")
```

```python
print("R2_Score:",r2_score(y_test,y_pred))
```

```python
r2 = r2_score(y_test,y_pred)
adj_r2 = 1-((1-r2)*(10-1)/(40-1-1))
print("Adjusted R2 Score:" , adj_r2)
```

```python
new_df1 = df.copy()
new_df1["random_features"] = np.random.random(200)
new_df1 = new_df1[["cgpa","random_features","package"]]
new_df1.head()
```

```python
plt.scatter(new_df1["random_features"] , new_df1["package"])
plt.xlabel("random_features")
plt.ylabel("package")
plt.title("random_features vs package")
```

```python
X = new_df1.drop("package" , axis = 1)
y = new_df1["package"]
```

```python
X_train , X_test ,y_train ,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
```

```python
print("R2_Score:",r2_score(y_test,y_pred))
```

```python
r2 = r2_score(y_test,y_pred)
adj_r2 = 1-((1-r2)*(10-1)/(40-1-1))
print("Adjusted R2 Score:" , adj_r2)
```
