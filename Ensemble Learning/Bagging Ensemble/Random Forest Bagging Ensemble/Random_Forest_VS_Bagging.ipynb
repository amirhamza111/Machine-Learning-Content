Random Forest Uses Different Features while creating new Decision Nodes but Bagging Use only selected Features while creating new nodes. This is the Major Difference B/w both. This Difference is explained below.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier , BaggingClassifier
from sklearn.datasets import make_classification
from sklearn.tree import plot_tree
```

```python
X,y = make_classification(n_features=5, n_redundant=0, n_informative=5,n_clusters_per_class=1)
```

```python

df = pd.DataFrame(X,columns=['col1','col2','col3','col4','col5'])
df['target'] = y
print(df.shape)
df.head()
```

# **Bagging**

```python
bag = BaggingClassifier(max_features=2)
```

```python
bag.fit(df.iloc[:,:-1],df.iloc[:,-1])
```

```python
plt.figure(figsize=(20,20))
plot_tree(bag.estimators_[0])
```

# **Random Forest**

```python
rf = RandomForestClassifier(max_features=2)
```

```python
rf.fit(df.iloc[:,:-1],df.iloc[:,-1])
```

```python
plt.figure(figsize=(20,20))
plot_tree(rf.estimators_[4])
```
