```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
```

```python
X , y = make_classification(n_features=5, n_redundant=0, n_informative=5,n_clusters_per_class=1)
```

```python
X
```

```python
y
```

```python
df = pd.DataFrame(X,columns=["col1","col2","col3","col4","col5"])
df["target"] = y
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(df,x="col1",y="col2",hue=df["target"],size=df["col3"])
```

# **Funtion for Row Sampling**

```python
def sample_rows(df,percent):
  return df.sample(int(percent*df.shape[0]),replace=True)

sample_rows(df,0.2)
```

# **Function for Feature Sampling**

```python
def sample_features(df,percent):
  import random
  cols = random.sample(df.columns.tolist()[:-1],int(percent*(df.shape[1]-1)))
  new_df = df[cols]
  new_df['target'] = df['target']
  return new_df

sample_features(df,0.4)
```

# **Function for Combined Sampling**

```python
def combined_sampling(df,row_per,col_per):
  new_df = sample_rows(df,row_per)
  return sample_features(new_df,col_per)

combined_sampling(df,0.5,0.4)
```

# **Random Forest with Row Sampling**

```python
df1 = sample_rows(df,0.5)
df2 = sample_rows(df,0.5)
df3 = sample_rows(df,0.5)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf1 = DecisionTreeClassifier()
clf2 = DecisionTreeClassifier()
clf3 = DecisionTreeClassifier()
```
