```python
# Data for Tree 1
df_bag1 = df_train.sample(50 , replace=True)
X_bag1 = df_bag1.drop('target',axis=1).sample(5,axis=1,replace=True)

y_bag1 = df_bag1['target']

dt_bag1 = DecisionTreeClassifier()
evaluate(dt_bag1,X_bag1,y_bag1)
```

```python
def evaluate(clf,X,y):
  clf.fit(X,y)
  plot_tree(clf)
  plt.show()
  y_pred = clf.predict(X)
  print(accuracy_score(y,y_pred))
```
