Ensemble Means Group of Musicians or Artists.

So Ensemble Learning Means That Group of Different Algorithm to train a Machine Learning Model.

Types of Ensemble Learning:-

1. Voting
2. Bagging (Random Forest)
3. Boosting ( AdaBoost , GradientBoost , XgBoost)
4. Stacking

1. Voting Ensemble (All the Base Models are Different)
2. Bagging Ensemble (Stands for BootStrap Aggregation , In this Base models are same but difference is that we provide different data to all the base models.)
3. Boosting Ensemble (It also has same base models. but each previous model tell about its mistakes to the next model. so the next model work well and learn from mistakes of previous one to boost its work.
4. Stacking Ensemble (Act same as Voting Ensenmble , means all base models are different but only difference is that it has an extra algorithm which trains on the data provided by base models.)
