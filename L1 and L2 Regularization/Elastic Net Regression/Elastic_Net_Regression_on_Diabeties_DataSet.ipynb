**Elastic Net Regression : It is the Combination of Lasso Regression and Ridge Regression.**

2025-04-22 11_05_31-(195) ElasticNet Regression _ Intuition and Code Example _ Regularized Linear Mo.png

**Loss Function of Elastic Net Regresssion:-**

L = (y-y_pred)^2 + lambda (w)^2 + lambda(abs(w))

Parameters:-
1. **Lambda** (Default Value is 1 )
2. **Ratio b/w a and b :** a/a+b (Default Values are a = 0.5 , b = 0.5)

```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression , Ridge , Lasso , ElasticNet
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
X , y = load_diabetes(return_X_y = True)
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,
                                                       test_size = 0.2,
                                                       random_state= 0)
```

```python
## Linear Regression
reg = LinearRegression()
reg.fit(X_train,y_train)
y_pred = reg.predict(X_test)
r2_score(y_test,y_pred)
```

```python
## Ridge Regression
reg = Ridge(alpha = 0.1)
reg.fit(X_train,y_train)
y_pred = reg.predict(X_test)
r2_score(y_test,y_pred)
```

```python
## Lasso
reg = Lasso(alpha = 0.01)
reg.fit(X_train,y_train)
y_pred = reg.predict(X_test)
r2_score(y_test,y_pred)
```

```python
## Elastic Net Regression
reg = ElasticNet(alpha = 0.005 , l1_ratio=0.9)
reg.fit(X_train,y_train)
y_pred = reg.predict(X_test)
r2_score(y_test,y_pred)
```

```python
## SGDRegressor
from sklearn.linear_model import SGDRegressor
reg = SGDRegressor(alpha = 0.001 , l1_ratio = 0.9)
reg.fit(X_train , y_train)
y_pred = reg.predict(X_test)
r2_score(y_test,y_pred)
```
