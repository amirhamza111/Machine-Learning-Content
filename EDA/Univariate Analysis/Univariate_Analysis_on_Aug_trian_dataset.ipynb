```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/train.csv")
df.head()
```

### **Categorical Data**

# a. CountPlot

```python
no , yes = df["Survived"].value_counts()
no , yes
```

```python
sns.countplot(x = "Survived", data = df)
plt.xlabel("Survived")
plt.ylabel("Count")
plt.show()
```

```python
plt.bar(x = ["Not Survived" , "Survived"] , height = [no , yes])
plt.show()
```

```python
df["Survived"].value_counts().plot(kind = "bar")
plt.show()
```

```python
df["Pclass"].value_counts().plot(kind = "bar")
plt.show()
```

```python
sns.countplot(x = "Pclass" , data = df)
```

```python
df["Sex"].value_counts().plot(kind = "bar")
plt.show()
```

```python
sns.countplot(x = "Embarked" , data = df)
```
