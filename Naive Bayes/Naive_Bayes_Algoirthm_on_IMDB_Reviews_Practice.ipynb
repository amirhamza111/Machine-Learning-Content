```python
import pandas as pd
import numpy as np
import tensorflow as tf
import sklearn
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("/content/movie_reviews (1).csv")
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df["Label"] = le.fit_transform(df["Label"])
df.head()
```

```python
X = df["Review"]
y = df["Label"]
```

```python
from sklearn.feature_extraction.text import CountVectorizer

v = CountVectorizer()
x_count = v.fit_transform(X.values)
x_count
```

```python
from sklearn.naive_bayes import MultinomialNB

m = MultinomialNB()
m.fit(x_count,y)
```

```python
# Generating a few sample movie reviews for prediction
sample = [
    "This movie was an absolute masterpiece, I loved every minute of it.",
    "The plot was confusing, and the characters were poorly developed.",
    "A thrilling action movie with amazing stunts and great visuals!",
    "This was the worst film I have ever watched, total waste of time.",
    "The acting was mediocre, but the story was heartwarming and touching.",
    "Boring and too slow, I wouldn't recommend it.",
    "An emotional rollercoaster with fantastic performances by the cast.",
    "The movie had some good moments, but overall it fell flat.",
]


def test(sample):
  sample_count = v.transform(sample)
  y_pred = m.predict(sample_count)
  x = ["Positive Review" if i == 1 else "Negative Review" for i in y_pred]
  return x

x = test(sample)
x
```
