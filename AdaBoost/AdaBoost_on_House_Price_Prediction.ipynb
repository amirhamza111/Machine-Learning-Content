```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/House Price Prediction Dataset.csv")
df.head()
```

```python
df1 = df.drop(["Id","Location","Condition"] , axis=1)
df1 = df1.replace(["No","Yes"],[0,1])
df1.head()
```

```python
df1.dtypes
```

```python
# from sklearn.preprocessing import MinMaxScaler
# s = MinMaxScaler()
# df2 = s.fit_transform(df1)
# df3 = pd.DataFrame(df2 , columns=df1.columns)
# df3.head()
```

```python
X = df1.drop("Price",axis=1)
y = df1["Price"]
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.2 , random_state=0)
```

```python
X_train.shape , X_test.shape , y_train.shape , y_test.shape
```

```python
from sklearn.ensemble import AdaBoostRegressor

model = AdaBoostRegressor(n_estimators=100 , random_state=0,loss="square")
model.fit(X_train  , y_train )
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred[:5]
```

```python
y_test.head()
```

```python
from sklearn.metrics import r2_score

r2_score(y_test,y_pred)
```
